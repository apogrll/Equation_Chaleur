```python
import numpy as np # bibliothèque pour la manipulation des vecteurs et matrices.
import matplotlib.pyplot as plt # fonctions de trace usuelles

# Pour tracer les figures à l'intérieur du notebook : pas de commentaires sur la ligne ci-dessous
%matplotlib inline
```

```python
L = 1. # longueur du domaine
T = 1.0 # Intervalle de temps
nu = 0.1 # coefficient de diffusion

# condition initiale
def u0(x):
    return 1. + np.sin(2*np.pi*x/L)
```

```python
# Test de la fonction initiale
u0(2)
```

```python
N = 100 # nombre d'intervalles (pas le nombre de points)
dx = float(L) / float(N)
alpha = 0.25 # alpha = nu*dt/dx^2 <= 1/2 pour la stabilité
dt = alpha*dx*dx/nu

xx = np.linspace(0,L,N+1)
```

```python
u = u0(xx)

# Affichage pour vérification
plt.plot(xx, u)
plt.show()
```

```python
# initialisation de la boucle en temps
time = 0.

# paramètres pour la gestion des affichages
N_affichage = 20
dt_affichage = T / N_affichage

# affichage de la condition initiale
print "affichage : temps = ",time," ", # on ajoute une virgule à la fin pour supprimer le retour à la ligne
plt.plot(xx, u)
t_affichage_suivant = time + dt_affichage
print "t_affichage_suivant = ",t_affichage_suivant

# boucle en temps
while (time < T):
    #print time
    
    # Conditions aux limites homogènes de Dirichlet
    u[1:N] = alpha*u[0:N-1] + (1-2*alpha)*u[1:N] + alpha*u[2:N+1]
    u[0] = 0
    u[N] = 0
    
    # Conditions périodiques : exercice
    
    
    # affichage ?
    if (time >= t_affichage_suivant):
        print time, 
        plt.plot(xx, u)

        t_affichage_suivant += dt_affichage # notez la syntaxe
        print "t_affichage_suivant = ",t_affichage_suivant



    # passage au pas de temps suivant    
    time = time + dt # on pourrait utiliser la syntaxe de la ligne précédente
    
# dernier affichage
print time
plt.plot(xx, u)
plt.show()
```
